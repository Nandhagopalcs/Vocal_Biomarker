```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import IPython.display as ipd
import librosa
import librosa.display
import os
import pickle
```

```python
file = open('data.pkl', 'rb')
data = pickle.load(file)
file.close()
```

```python
covid_positive = data[data['COVID_test_status'] ==1]
```

```python
covid_negative = data.head(11)
```

```python
data=pd.concat([covid_positive,covid_negative])
```

```python
data['COVID_test_status'].value_counts()
```

```python
data.reset_index(drop=True, inplace=True)
```

```python
y=data[['COVID_test_status']].copy()
data.drop('COVID_test_status', axis=1, inplace=True)
```

```python
data.head()
```

```python
y.head()
```

```python
breathing_deep=np.array(data['breathing-deep'].tolist())
breathing_shallow=np.array(data['breathing-shallow'].tolist())
cough_heavy=np.array(data['cough-heavy'].tolist())
cough_shallow=np.array(data['cough-shallow'].tolist())
counting_fast=np.array(data['counting-fast'].tolist())
counting_normal=np.array(data['counting-normal'].tolist())
vowel_a	=np.array(data['vowel-a'].tolist())
vowel_e	=np.array(data['vowel-e'].tolist())
vowel_o=np.array(data['vowel-o'].tolist())
```

```python
x=np.concatenate([breathing_deep,breathing_shallow,cough_heavy,cough_shallow,counting_fast,counting_normal,vowel_a,vowel_e,vowel_o],axis=1)
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

## Model

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
num_labels=y.shape[1]
```

```python
model=Sequential()
###first layer
model.add(Dense(360,input_shape=(360,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###second layer
model.add(Dense(200))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###third layer
model.add(Dense(100))
model.add(Activation('relu'))
model.add(Dropout(0.5))

###final layer
model.add(Dense(num_labels))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',metrics=['accuracy'],optimizer='adam')
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
from datetime import datetime 

num_epochs = 100
num_batch_size = 32

checkpointer = ModelCheckpoint(filepath='saved_models/audio_classification.hdf5', 
                               verbose=1, save_best_only=True)
start = datetime.now()

model.fit(X_train, y_train, batch_size=num_batch_size, epochs=num_epochs, validation_data=(X_test, y_test), callbacks=[checkpointer], verbose=1)


duration = datetime.now() - start
print("Training completed in time: ", duration)
```
